```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm as cm
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
centers = [(-5, -5), (1, -2), (5, 5),(12, 7)]
X, y = make_blobs(100, 2, centers=centers, random_state=2, cluster_std=1.5)

fig, ax = plt.subplots()

ax.scatter(X[:, 0], X[:, 1], c=y, s=14, cmap='RdBu')
ax.set_title('Naive Bayes Model', size=14)

cmap = sns.color_palette('RdBu',len(centers))


xlim = (-12, 17)
ylim = (-15, 15)

xg = np.linspace(xlim[0], xlim[1], 60)
yg = np.linspace(ylim[0], ylim[1], 40)
xx, yy = np.meshgrid(xg, yg)
Xgrid = np.vstack([xx.ravel(), yy.ravel()]).T

for label, color in enumerate(['red','red','blue','blue']):
    mask = (y == label)
    mu, std = X[mask].mean(0), X[mask].std(0)
    P = np.exp(-0.5 * (Xgrid - mu) ** 2 / std ** 2).prod(1)
    Pm = np.ma.masked_array(P, P < 0.03)
    ax.pcolorfast(xg, yg, Pm.reshape(xx.shape), alpha=0.5,
                  cmap=color.title() + 's')
    ax.contour(xx, yy, P.reshape(xx.shape),
               levels=[0.01, 0.1, 0.5, 0.9],
               colors=cmap[label], alpha=0.2)
    
ax.set(xlim=xlim, ylim=ylim)
ax.set_xlabel("Neuron 1 Activity", size=14)
ax.set_ylabel("Neuron 2 Activity", size=14)
```
